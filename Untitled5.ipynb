##Problem Statement 1:
Prepare a dashboard for the all the Accounts doing Credit transactions from Moravia and Prague. preparing the aggregated view of the transactions basis on 2 things i.e.

1 - Account Wise 2 - Month Wise 3 - Account and Month wise.

```python
df7.head(5)
```

```python
df7['Month'] = df7['date_y'].dt.month
```

```python
# Create a function to generate account-wise and month-wise aggregated views
def generate_dashboard():
    # Account-wise view
    account_wise_view = df7.groupby('account_id')['amount'].sum().reset_index()

    # Month-wise view
    month_wise_view = df7.groupby('Month')['amount'].sum().reset_index()

    # Account and Month-wise view
    account_month_wise_view = df7.groupby(['account_id', 'Month'])['amount'].sum().reset_index()

    return account_wise_view, month_wise_view, account_month_wise_view
```

```python
# Generate the views
account_wise_view, month_wise_view, account_month_wise_view = generate_dashboard()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create visualizations
plt.figure(figsize=(15, 5))

# Account-wise view
plt.subplot(131)
sns.barplot(data=account_wise_view, x='account_id', y='amount')
plt.title('Account-wise Credit Transactions')
plt.xlabel('Account')
plt.ylabel('Total Amount')

# Month-wise view
plt.subplot(132)
sns.barplot(data=month_wise_view, x='Month', y='amount')
plt.title('Month-wise Credit Transactions')
plt.xlabel('Month')
plt.ylabel('Total Amount')

# Account and Month-wise view
plt.subplot(133)
sns.heatmap(pd.pivot_table(account_month_wise_view, values='amount', index='account_id', columns='Month'), annot=True, cmap='YlGnBu')
plt.title('Account and Month-wise Credit Transactions')
plt.xlabel('Month')
plt.ylabel('Account')

plt.tight_layout()
plt.show()
```

##Problem Statement 3:
Number of cards issued in district where average salary is more than 9000, is it a good strategy?

```python
# Calculate average salary per district
district_avg_salary = df7.groupby('district_id')['A11'].mean()  # Assuming 'A11' is the column representing average salary

# Count the number of unique card IDs issued in each district
district_card_counts = df7.groupby('district_id')['card_id'].nunique()

# Merge the data
district_data = pd.merge(district_avg_salary, district_card_counts, on='district_id', how='inner')
district_data.columns = ['avg_salary', 'num_cards_issued']

# Filter districts with average salary > 9000
high_salary_districts = district_data[district_data['avg_salary'] > 9000]

# Calculate correlation
correlation = high_salary_districts['avg_salary'].corr(high_salary_districts['num_cards_issued'])

print("Correlation between average salary and number of cards issued:", correlation)
```

##Problem Statement 4:
Are we providing loans to members belonging to district where committed crimes are more than 6000 for code 95, if yes then provide the number of loans per district.

```python
df7['A12'] = df7['A12'].astype(str)
```

```python
# Convert 'A12' column values to floats
df7['A12'] = df7['A12'].str.replace(',', '.').astype(float)

# Filter districts where the unemployment rate for any year is greater than 2%
high_unemployment_districts = df7[df7.groupby('district_id')['A12'].transform('max') > 2]  # Assuming 'A12' is the unemployment rate column

# Sum the 'amount_y' column for customers in these districts
total_collected_from_other_banks = high_unemployment_districts['amount_y'].sum()

print("Total money collected from other banks for customers in high unemployment districts:", total_collected_from_other_banks)
```

##Problem Statement 5:
How many customers having credit card are also availing the loan facilities.

```python
# Filter customers with credit cards
credit_card_customers = df7[(df7['type_y'] == 'gold') | (df7['type_y'] == 'classic')]  # Assuming 'type_y' is the card type column

# Filter customers with loan facilities
loan_customers = df7[df7['A12'] > 0]  # Assuming 'A12' is the loan column

# Merge the two filtered datasets to find customers with both credit cards and loans
customers_with_both = pd.merge(credit_card_customers, loan_customers, on='client_id_y', how='inner')

# Count the number of customers who have both credit cards and loans
num_customers_with_both = customers_with_both['client_id_y'].nunique()

print("Number of customers having credit cards and availing loan facilities:", num_customers_with_both)
```

##Problem Statement 6:
Can we say that customers having Classic and Junior card are the ones who are more in debt.

```python
# Filter customers with Classic and Junior cards
classic_junior_customers = df7[(df7['type_y'] == 'classic') | (df7['type_y'] == 'junior')]  # Assuming 'type_y' is the card type column

# Calculate the average loan amount for each card type
average_loan_amounts = classic_junior_customers.groupby('type_y')['A12'].mean()  # Assuming 'A12' is the loan column

# Compare the average loan amounts to draw conclusions
average_loan_classic = average_loan_amounts.get('classic', 0)
average_loan_junior = average_loan_amounts.get('junior', 0)

if average_loan_classic > average_loan_junior:
    print("Customers with Classic cards are more in debt.")
elif average_loan_classic < average_loan_junior:
    print("Customers with Junior cards are more in debt.")
else:
    print("Customers with Classic and Junior cards have similar debt levels.")
```

##Problem Statement 7:
How will you analyze the performance of Mid age vs adults in terms of loan repayments

```python
# Convert birth_number to datetime to calculate age
df7['birth_number'] = pd.to_datetime(df7['birth_number'], format='%y%m%d')

# Calculate age based on birth_number
current_year = pd.Timestamp.today().year
df7['age'] = current_year - df7['birth_number'].dt.year

# Categorize customers into mid-age and adult groups (adjust the age thresholds as needed)
mid_age_customers = df7[(df7['age'] >= 30) & (df7['age'] <= 50)]
adult_customers = df7[df7['age'] > 50]

# Calculate loan repayment rates for both groups
mid_age_repayment_rate = mid_age_customers['status'].value_counts(normalize=True).get('A', 0)
adult_repayment_rate = adult_customers['status'].value_counts(normalize=True).get('A', 0)

# Compare and analyze the loan repayment performance
print("Loan Repayment Performance Analysis:")
print("Mid-Age Customers (30-50 years old) - Repayment Rate:", mid_age_repayment_rate)
print("Adult Customers (>50 years old) - Repayment Rate:", adult_repayment_rate)

if mid_age_repayment_rate > adult_repayment_rate:
    print("Mid-age customers have better loan repayment performance.")
elif mid_age_repayment_rate < adult_repayment_rate:
    print("Adult customers have better loan repayment performance.")
else:
    print("Loan repayment performance is similar for mid-age and adult customers.")
```
